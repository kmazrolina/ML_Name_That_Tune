# Christmas Carol Melody Classification

### Import and preparation of csv data

```python
import numpy as np
import pandas as pd
import os
import os
import pandas as pd
    
# Specify the directory containing your CSV files
directory_path = f'koledy/note_sequences'

# Get a list of all CSV files in the directory
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]
csv_files.sort()

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through each CSV file and read it into a DataFrame
for csv_file in csv_files:
    file_path = os.path.join(directory_path, csv_file)
    print(file_path)
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Append the DataFrame to the list
    dfs.append(df)

s_id = 3 #start index for my datasets in array
```

#### Initial look at the data

```python
def midi_to_classic_interval(midi_value):
    if midi_value >= 0:
        plus_sign = "+"
    else:
        plus_sign = ""
    for classic_value in range (8, 1, -1):
        if midi_value % classic_value == 0:
            return "{} {}{} octaves".format(classic_value,plus_sign, midi_value/12)
```

```python
print("Intervals")
print("\tBiggest downwards melody jump: ", midi_to_classic_interval(dfs[7 - s_id].filter(regex="Interval[0-9]").min().min()))
print("\tBiggest upwards melody jump: ", midi_to_classic_interval(dfs[7 - s_id].filter(regex="Interval[0-9]").max().max()))

print("Rythmic_values")
print("\tLongest rythmical value: ", dfs[7 - s_id].filter(regex="Rythmic_Value[0-9]").max().max())
print("\tShortest rythmical value: ", dfs[7 - s_id].filter(regex="Rythmic_Value[0-9]").min().min())

print("KolÄ™dy")
dfs[7 - s_id]['Koleda'].unique()
```
